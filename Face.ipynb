```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
frame = Image.open("Tirth.jpg")
bgr = cv2.imread("Tirth.jpg")
```

```python
plt.imshow(bgr)
```

```python
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(rgb)
```

```python
face = faces[0]
```

```python
x,y,w,h = face
```

```python
FACE = rgb[y:y+h,x:x+w]
```

```python
plt.imshow(FACE)
```

```python
outline = cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,0,255),6)
```

```python
plt.imshow(outline)
```
